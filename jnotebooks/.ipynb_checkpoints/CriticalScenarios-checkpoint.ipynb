Loads results from exporting critical scenarios and the collision risk estimations.

```julia
using AutomotiveDrivingModels
using AutoScenes
using AutoViz
using Records
using PGFPlots

include("../AutoMobius/AutoMobius.jl")
using AutoMobius

pushPGFPlotsPreamble("\\usepackage{amsmath}")

const DATA_DIR = "/media/tim/DATAPART1/PublicationData/2017_adas_validation/"
const OUTPUT_FILE_SUFFIX = "1010"
const COLLISION_SCENARIO_DIR = joinpath(DATA_DIR, "collision_scenarios_$OUTPUT_FILE_SUFFIX")

"DONE"
```

### load data

```julia
packedscenes = open(joinpath(COLLISION_SCENARIO_DIR, "crit_transition_scenes.txt"), "r") do io
    read(io, MIME"text/plain"(), Vector{AttentiveMobiusScene})
end

struct CritSceneBlame
    rear::Int
    fore::Int
    local_tick::Int
    global_tick::Int
    filename::String
end
blames = open(joinpath(COLLISION_SCENARIO_DIR, "crit_transition_blames.txt"), "r") do io
    retval = Array{CritSceneBlame}(length(packedscenes))
    lines = readlines(io)
    for i in 1 : length(retval)
        line = lines[i]
        tokens = split(strip(line), ",")
        filename=strip(tokens[1])
        rear = parse(Int, tokens[2])
        fore = parse(Int, tokens[3])
        local_tick = parse(Int, tokens[4])
        global_tick = parse(Int, tokens[5])
        retval[i] = CritSceneBlame(rear, fore, local_tick, global_tick,filename)
    end
    retval
end

indices = [b.rear != 0 for b in blames]
blames = blames[indices]
packedscenes = packedscenes[indices]

"DONE"
```

```julia
struct RiskAssessment
    nsims::Int
    nticks::Int
    ncol::Int
    est_risk::Float64
end
risk_curves = Array{Vector{RiskAssessment}}(length(blames))
for (j,b) in enumerate(blames)
    MC_risk_path = splitext(b.filename)[1] * "_risks.txt"
    lines = readlines(joinpath(COLLISION_SCENARIO_DIR, MC_risk_path))
    mc_risks = Array{RiskAssessment}(length(lines))
    for (i,line) in enumerate(lines)
        tokens = split(strip(line), ",")
        mc_risks[i] = RiskAssessment(
            parse(Int, tokens[1]),
            parse(Int, tokens[2]),
            parse(Int, tokens[3]),
            parse(Float64, tokens[4]),
        )
    end
    risk_curves[j] = mc_risks
end
```

```julia
plots = Plots.Plot[]

function get_risk(r::RiskAssessment)
    mean(Beta(1 + r.ncol, 1 + r.nsims - r.ncol))
end

max_n_curves_to_plot = 100
min_time = 0.0
risk_threshold = 0.15

for curve in risk_curves
    if get_risk(curve[end]) > 0.8
        max_n_curves_to_plot -= 1
        risks = [get_risk(r) for r in curve]
        times = (collect(1:length(curve)) - length(curve)).*timestep
        push!(plots, Plots.Linear(times, risks, style="solid, ultra thick, blue, opacity=0.2, mark=none"),)
        min_time = min(min_time, minimum(times))
        if max_n_curves_to_plot <= 0
            break
        end
    end
end

push!(plots, 
    Plots.Linear([min_time,0], [risk_threshold,risk_threshold], style="solid, black, mark=none")
)

ax = Axis(plots, style="axis on top, enlarge x limits=0, ymin=0, ymax=1, xmin=-20", xlabel="time [s]", ylabel="estimated risk")
# PGFPlots.save("/home/tim/Documents/papers/2017_thesis/figures/risk_curve.tex", ax, include_preamble=false)
ax
```

```julia
mutable struct CriticalCarOverlay <: SceneOverlay
    target_id::Int
end
function AutoViz.render!(rendermodel::RenderModel, overlay::CriticalCarOverlay, scene::MobiusScene, roadway::MobiusRoadway)
    ind = findfirst(scene, overlay.target_id)
    if ind > 0
        veh = scene[ind]
        pts = Matrix{Float64}(2,2)
        pts[1,:] = veh.state.s
        pts[2,1] = -12
        pts[2,2] = -5
        add_instruction!(rendermodel, render_arrow, (pts, colorant"white", 0.25, 3.0))
    end
    return rendermodel
end
```

```julia
crit_scenes = [unpack_critical_scene!(MobiusScene(), list_records[k], roadway, risk_ests[k], risk_threshold) for k in lrec_keys]
"DONE"
```

```julia
scene_sizes = [length(s) for s in crit_scenes]
Axis(Plots.Histogram(scene_sizes, bins=20, discretization=:specified), 
     ymin=0, xlabel="number of cars", ylabel="counts")
```

```julia
import QuadGK: quadgk

critical_frame_indices = [findfirst_frame_above_risk_threshold(risk_ests[k], risk_threshold) for k in lrec_keys]

N = length(lrec_keys)
critical_cars = Array{Tuple{Int,Int}}(N)
delta_s_crit = Array{Float64}(N)
delta_v_crit = Array{Float64}(N)
for (idx, k) in enumerate(lrec_keys)
    scene_end = unpack!(list_records[k], nframes(list_records[k]), MobiusScene())
    i,j = get_first_collision(scene_end, roadway)
    scene = crit_scenes[idx]
    delta_s_ij = get_headway(scene[i], scene[j], roadway)
    delta_s_ji = get_headway(scene[j], scene[i], roadway)
    if delta_s_ij < delta_s_ji
        critical_cars[idx] = (i,j)
        delta_s_crit[idx] = delta_s_ij
        delta_v_crit[idx] = scene[j].state.v - scene[i].state.v
    else
        critical_cars[idx] = (j,i)
        delta_s_crit[idx] = delta_s_ji
        delta_v_crit[idx] = scene[i].state.v - scene[j].state.v
    end
end


frame = EntityFrame(AttentiveMobiusState,AttentiveMobiusDef,Int,100)
started_off_critical = critical_frame_indices .== 1

timestep = 0.1
P_seconds_to_attentive = LogNormal(0.6133, 0.8934)
P_seconds_to_distracted = LogNormal(3.2817, 0.7516)

true_time_to_collision = Vector{Float64}(N)
is_attentive_rear = Vector{Bool}(N)
is_attentive_front = Vector{Bool}(N)
expected_time_to_inattention_rear = Vector{Float64}(N)
expected_time_to_inattention_front = Vector{Float64}(N)
for (i,k) in enumerate(lrec_keys)
    scene = get!(frame, list_records[k], critical_frame_indices[i])
    
    rear = scene[critical_cars[i][1]]
    fore = scene[critical_cars[i][2]]
    is_attentive_rear[i] = rear.state[2].is_attentive
    is_attentive_front[i] = fore.state[2].is_attentive
    true_time_to_collision[i] = timestep*(nframes(list_records[k]) - critical_frame_indices[i])
    
    function expected_time_to_inattention(s)
        if !s.is_attentive
            return 0.0
        else
            t_lo = s.steps_since_attention_swap * timestep
            return quadgk(t->t*pdf(P_seconds_to_distracted, t+t_lo), 0.0, Inf)[1] / (1 - cdf(P_seconds_to_distracted, t_lo))
        end
    end
    
    expected_time_to_inattention_rear[i] = expected_time_to_inattention(rear.state[2])
    expected_time_to_inattention_front[i] = expected_time_to_inattention(fore.state[2])
end

using DataFrames
using PGFPlots
using PGFBarsAndPies

df = DataFrame()
df[:scene_sizes] = scene_sizes
df[:started_off_critical] = [c ? "crit" : "benign" for c in started_off_critical]
df[:is_attentive_rear] = [b ? "attentive" : "distracted" for b in is_attentive_rear]
df[:is_attentive_front] = [b ? "attentive" : "distracted" for b in is_attentive_front]
df[:true_time_to_collision] = true_time_to_collision
df[:delta_s_crit] = delta_s_crit
df[:delta_v_crit] = delta_v_crit
df[:ttc_crit] = [v < 0 ? min(-s/v,30.0) : 40.0 for (s,v) in zip(delta_s_crit, delta_v_crit)]
df[:expected_time_to_inattention_rear] = expected_time_to_inattention_rear
df[:expected_time_to_inattention_front] = expected_time_to_inattention_front

function plotdf(df)
    g = GroupPlot(3,4, groupStyle="horizontal sep=1.5cm, vertical sep=2cm")
    push!(g, Axis(BarChart(df[:started_off_critical]), xlabel="initialization"))
    push!(g, Axis(BarChart(df[:is_attentive_rear]), xlabel="rear state"))
    push!(g, Axis(BarChart(df[:is_attentive_front]), xlabel="fore state"))
    push!(g, Axis(Plots.Histogram(convert(Vector{Float64}, df[:scene_sizes])), ymin=0, xlabel="scene size", style="axis on top"))
    push!(g, Axis(Plots.Histogram(convert(Vector{Float64}, df[:delta_s_crit])), ymin=0, xlabel=L"\Delta s_\text{crit}", style="axis on top"))
    push!(g, Axis(Plots.Histogram(convert(Vector{Float64}, df[:delta_v_crit])), ymin=0, xlabel=L"\Delta v_\text{crit}", style="axis on top"))
    push!(g, Axis(Plots.Histogram(convert(Vector{Float64}, df[:ttc_crit])), ymin=0, xlabel=L"ttc_\text{crit}", style="axis on top"))
    push!(g, Axis(Plots.Histogram(convert(Vector{Float64}, df[:expected_time_to_inattention_rear]), discretization=:specified, bins=10), ymin=0, xlabel="expected time to inattention, rear", style="axis on top"))
    push!(g, Axis(Plots.Histogram(convert(Vector{Float64}, df[:expected_time_to_inattention_front]), discretization=:specified, bins=10), ymin=0, xlabel="expected time to inattention, fore", style="axis on top"))
    push!(g, Axis(Plots.Histogram(convert(Vector{Float64}, df[:true_time_to_collision])), ymin=0, xlabel="true time to collision", style="axis on top"))
end

plotdf(df)
```
