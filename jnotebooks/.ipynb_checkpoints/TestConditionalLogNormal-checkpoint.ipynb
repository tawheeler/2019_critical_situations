# Test Conditional LogNormal

```julia
using Distributions
using PGFPlots
```

```julia
P = LogNormal(0.6133, 0.8934)
Q = LogNormal(3.2817, 0.7516)

Axis([
Plots.Linear(x->pdf(P, x), (0,100), style="blue"),
Plots.Linear(x->pdf(Q, x), (0,100), style="red"),
        ], xlabel="number of ticks before transition", ylabel="likelihood")
```

```julia
function get_probability_of_transition_at(P::LogNormal, t_lo::Float64, t_hi::Float64)
    # probability of transition in t_lo : t_hi given that
    # it did not occur in 0 : t_lo
    cdf_lo = cdf(P, t_lo)
    cdf_hi = cdf(P, t_hi)
    return (cdf_hi - cdf_lo) / (1 - cdf_lo)
end

function get_time_to_transition(P::LogNormal, dt::Float64 = 0.1)
    t = 0.0
    while rand() > get_probability_of_transition_at(P, t, t+dt)
        t += dt
    end
    t += rand()*dt
    return t
end
```

```julia
samples = [get_time_to_transition(Q, 0.1) for i in 1 : 10000]


Axis([Plots.Histogram(samples, density=true),
      Plots.Linear(x->pdf(Q, x), (0,maximum(samples)), style="red"),
        ], ymin=0, xlabel="time [s]", ylabel="likelihood")
```
