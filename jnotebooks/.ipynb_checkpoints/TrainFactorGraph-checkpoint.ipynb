```julia
factorgraphs = [FactorGraph(features, scene, roadway) for scene in scenes]
θ = ones(length(features))
∇ = Array{Float64}(length(θ))
srand(0)
@time log_pseudolikelihood_gradient!(∇, features, θ, factorgraphs[1:10])
@time log_pseudolikelihood_gradient!(∇, features, θ, factorgraphs[1:10])
```

```julia
epoch = length(factorgraphs)

θ = ones(length(features))
∇ = Array{Float64}(length(θ))
sampler = BatchIterator(epoch)
batch = Array{eltype(factorgraphs)}(1)
b = i -> clamp(round(Int, sqrt(i)), 1, epoch)
logPLs = Float64[]
iters = Int[]
θs = Vector{Float64}[]
iter = 0

push!(logPLs, log_pseudolikelihood(features, θ, factorgraphs))
push!(iters, 1)

srand(0)
"READY"
```

```julia
α = 0.01
λ = 0.01 # regularization term
m = 2000
∇max = 1.0
skip = 20

tic()

for i in 1 : m
    iter += 1
    resize!(batch, b(iter))
    pull_batch!(batch, factorgraphs, sampler)
    log_pseudolikelihood_gradient!(∇, features, θ, batch)
    ∇ -= 2λ*θ
    θ += (x->clamp(x, -∇max, ∇max)).(α*∇)
    if mod(iter, skip) == 0
        push!(logPLs, log_pseudolikelihood(features, θ, factorgraphs))
        push!(iters, iter)
        push!(θs, deepcopy(θ))
    end
end
toc()

"DONE"
```

```julia
Axis(Plots.Linear(1:length(logPLs), logPLs, style="solid, blue, mark=*"),
    xlabel="iteration", ylabel="log pseudolikelihood loss")
```

```julia
p = Plots.Plot[]
colors = ["red", "green", "blue", "magenta", "cyan", "orange", "red!50", "green!50", "blue!50", "magenta!50", "cyan!50"]
x_arr = collect(1:length(θs)) .- 1
for i in 1 : length(θs[1])
    weights = [θ[i] for θ in θs]
    push!(p, Plots.Linear(x_arr, weights, style="solid, thick, $(colors[i]), mark=none", legendentry=string(features[i])))
end
Axis(p, xlabel="iteration", ylabel=L"\theta", width="23cm", height="8cm", style="legend pos=outer north east, enlarge x limits=0")
```
