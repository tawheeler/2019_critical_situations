# Passive Aggressive

```julia
using AutomotiveDrivingModels
if !isdefined(:AutoViz)
    include(Pkg.dir("AutomotiveDrivingModels", "src", "viz", "AutoViz.jl"))
end
using AutoViz

using Reactive, Interact, Reel

srand(0)
roadway = gen_stadium_roadway(3, radius=45.0)

start_roadinds = [RoadIndex(proj(VecSE2(0.0,-1*DEFAULT_LANE_WIDTH,0.0), roadway)),
                  RoadIndex(proj(VecSE2(0.0,-0*DEFAULT_LANE_WIDTH,0.0), roadway)),
                  RoadIndex(proj(VecSE2(0.0,-2*DEFAULT_LANE_WIDTH,0.0), roadway)),
                 ]
start_delta_s = 10.0:50.0:480.0
ncars = length(start_roadinds) * length(start_delta_s)

function gen_aggressive(action_context)
    mlon = IntelligentDriverModel(
        σ=0.1,
        k_spd=1.5,
        T=0.5,
        s_min=4.0,
        a_max=4.0,
        d_cmf=2.5,
        )
    mlat = ProportionalLaneTracker(σ=0.1, kp=3.5, kd=2.5)
    mlane = TimLaneChanger(action_context,
        threshold_fore = 50.0,
        threshold_lane_change_gap_fore = 5.0,
        threshold_lane_change_gap_rear = 5.0,
        )
    
    model = Tim2DDriver(context, rec=SceneRecord(1, context.Δt, ncars), mlat=mlat, mlon=mlon, mlane=mlane)
    set_desired_speed!(model, 30.0 + randn())
    model
end
function gen_passive(action_context)
    mlon = IntelligentDriverModel(
        σ=0.1,
        k_spd=1.0,
        T=1.75,
        s_min=5.0,
        a_max=3.0,
        d_cmf=2.0,
        )
    mlat = ProportionalLaneTracker(σ=0.1, kp=3.0, kd=2.0)
    mlane = TimLaneChanger(action_context,
        threshold_fore = 100.0,
        threshold_lane_change_gap_fore = 10.0,
        threshold_lane_change_gap_rear = 10.0,
        )
    
    model = Tim2DDriver(context, rec=SceneRecord(1, context.Δt, ncars), mlat=mlat, mlon=mlon, mlane=mlane)
    set_desired_speed!(model, 20.0 + randn())
    model
end

scene = Scene()
framerate = 24
context = IntegratedContinuous(1/framerate,3)
models = Dict{Int, DriverModel}()
carcolors = Dict{Int,Colorant}()
base_speed = 20.0
x = 10
j = 0
id_count = 0
for roadind in start_roadinds
    vehstate = VehicleState(Frenet(roadind, roadway), roadway, base_speed+randn())
    for delta_s in start_delta_s    
        vehstate2 = move_along(vehstate, roadway, delta_s)
        vehdef = VehicleDef(id_count+=1, AgentClass.CAR, 4.826, 1.81)
        push!(scene,Vehicle(vehstate2, vehdef))
    
        if rand() > 0.5
            models[id_count] = gen_aggressive(context)
            carcolors[id_count] = colorant"red"
        else
            models[id_count] = gen_passive(context)
            carcolors[id_count] = colorant"blue"
        end
    end
end

actions = get_actions!(Array(DriveAction, length(scene)), scene, roadway, models)

overlays = SceneOverlay[] 
# overlays = [CarFollowingStatsOverlay(1)] 
# overlays = [NeighborsOverlay(2)]

function drawsim(t, dt=NaN)
    get_actions!(actions, scene, roadway, models)
    tick!(scene, roadway, actions, models)
    render(scene, roadway, overlays, cam=FitToContentCamera(), special_car_colors=carcolors)
end

ticks = fps(framerate)
timestamps = map(_ -> time(), ticks)
map(drawsim, timestamps)

# while true
#     drawsim(1)
# end

# film = roll(drawsim, fps=framerate, duration=20)
# write("tim2d_05.gif", film) # Write to a gif file
```
