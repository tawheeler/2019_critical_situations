```julia
roadway = Wraparound(Straight1DRoadway(200.0))
scenes_crit = open("../data/crit_scenes_0731.txt") do io
     read(io, MIME"text/plain"(), Vector{MobiusScene})
end
# scenes_fact = open("../data/scenes_attentivemobiusfactorgraph_0728.txt") do io
#      read(io, MIME"text/plain"(), Vector{MobiusScene})
# end
critical_cars = open("../data/crit_cars_0731.txt") do io
    lines = readlines(io)
    n = parse(Int, strip(lines[1]))
    critical_cars = Vector{Vector{Int}}(n)
    for i in 1 : n
        tokens = split(strip(lines[i+1]))
        critical_cars[i] = parse.(Int, tokens)
    end
    critical_cars
end

scenes_crit = scenes_crit[1001:end] # test
crit_cars_crit = critical_cars[1001:end] # test
crit_cars_fact = critical_cars[1:100] # train
"done"
```

```julia
include("../crit_cluster_sampling/critical_scenario_metrics.jl")
critscenes_crit = [MobiusCritScene(s, c) for (s,c) in zip(scenes_crit, crit_cars_crit)]
"DONE"
```

```julia
metrics = Dict(
            "v"      => KLDivMetric( v,      LinearDiscretizer(linspace(-10.0, 35.0, 41))),
           "Δv"      => KLDivMetric(Δv,      LinearDiscretizer(linspace(-15.0, 20.0, 41))),
           "Δs"      => KLDivMetric(Δs,      LinearDiscretizer(linspace(0.0, 100.0, 41))),
           "Δv_crit" => KLDivMetric(Δv_crit, LinearDiscretizer(linspace(-15.0, 20.0, 41))),
           "Δs_crit" => KLDivMetric(Δs_crit, LinearDiscretizer(linspace(0.0, 100.0, 41))),
          )
countsP = Dict(k=>get_counts(v, critscenes_crit, roadway) for (k,v) in metrics)
# countsQ = get_counts(metric, scenes_fact, roadway) .+ 1
# kldiv = kullbeck_leibler_divergence(countsP, countsQ, metric.disc)
```

```julia
function counts_line_plot(counts::Vector{Int}, disc::LinearDiscretizer, style::String="")
    arr_x, arr_y = Float64[], Float64[]
    tot_count = sum(countsP)
    for (bin, count) in enumerate(counts)
        lo, hi = extrema(disc, bin)
        val = count/tot_count
        append!(arr_x, [lo, hi])
        append!(arr_y, [val, val])
    end
    return Plots.Linear(arr_x, arr_y, style=style)
end
```
