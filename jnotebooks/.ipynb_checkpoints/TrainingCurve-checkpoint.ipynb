```julia
using PGFPlots
```

```julia
include("parse_training_curve.jl")
# curve = parse_training_curve("test.txt")
curve = parse_training_curve("../trainout.txt")

g = GroupPlot(2, 2, groupStyle = "horizontal sep = 2cm, vertical sep = 1.5cm")
push!(g, Axis(Plots.Linear(curve.time, curve.ave_reward, style="mark=none, blue"),
         xlabel="time [min]", ylabel="average reward"))
push!(g, Axis(Plots.Linear(curve.time, curve.kldiv, style="mark=none, orange"),
         xlabel="time [min]", ylabel="KL-divergence"))
push!(g, Axis(Plots.Linear(curve.time, curve.surrogate_loss, style="mark=none, red"),
         xlabel="time [min]", ylabel="surrogate loss"))
push!(g, Axis(Plots.Linear(curve.time, curve.entropy, style="mark=none, green"),
         xlabel="time [min]", ylabel="entropy"))
g
```
