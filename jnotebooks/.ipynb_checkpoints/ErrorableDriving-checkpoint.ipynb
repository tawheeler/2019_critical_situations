This notebook shows how to run simulations with the errorable driver model.

```julia
using AutomotiveDrivingModels
# using AutoRisk
using AutoViz
using Records
using HDF5, JLD
using PGFPlots
using BenchmarkTools

include("../AutoMobius/AutoMobius.jl")
using AutoMobius

pushPGFPlotsPreamble("\\usepackage{amsmath}")
"DONE"
```

```julia
srand(0)
timestep = 0.1
scenegen = BayesNets1DSceneModel()
scene = MobiusScene()
roadway = Wraparound(Straight1DRoadway(200.0))
models = Dict{Int, LaneFollowingDriver}()
behgen = ErrorableIDMBehaviorGenerator(timestep)

rand!(scene, scenegen, roadway)
rand(models, behgen, scene)

cam = StaticCamera(VecE2(100.0,0.0), 4.75)
car_colors = get_pastel_car_colors(scene)
render(scene, roadway, [AttentionOverlay(id, model) for (id,model) in models], 
       cam=cam, canvas_height=100, car_colors=car_colors)
```

Run a simulation until collision.

```julia
record_history = 100
rec = QueueRecord(MobiusEntity, record_history, timestep)
driverstate_rec = QueueRecord(AttentiveEntity, record_history, timestep)
simscene = copy(scene)

rec, driverstate_rec = simulate_until_collision(copy!(simscene, scene), roadway, models, rec, driverstate_rec, nticks=100000)
render(rec[0], roadway,[AttentionOverlay(id, model) for (id,model) in models], cam=cam, canvas_height=100, car_colors=car_colors)
```

Get collision estimation, which is a Beta belief over collision occuring.
Number of simulations can be set with `nsimulations` variable.

```julia
list_record = pack_output_list_record(rec, driverstate_rec, models)
prior = 0.1
get_collision_estimation(list_record, 5, roadway, prior, nsimulations=10)
```

View the sim.

```julia
using Interact
@manipulate for frame_index in 1 : nframes(list_record)
    AutoMobius.unpack!(list_record, frame_index, simscene, models)
    render(simscene, roadway,[AttentionOverlay(id, model) for (id,model) in models], cam=cam, canvas_height=100, car_colors=car_colors)
end
```

```julia
AutoMobius.unpack!(list_record, nframes(list_record), simscene, models)
col = get_first_collision(simscene, roadway)
target_ind = col[1]

@manipulate for frame_index in 1 : nframes(list_record)
    AutoMobius.unpack!(list_record, frame_index, simscene, models)
    
    # adjust target car to be at center
    del_s = simscene[target_ind].state.s + 100.0
    for i in 1 : length(simscene)
        veh = simscene[i]
        s′ = mod_position_to_roadway(veh.state.s - del_s, roadway)
        simscene[i] = Entity(PosSpeed1D(s′, veh.state.v), veh.def, veh.id)
    end
    
    render(simscene, roadway,[AttentionOverlay(id, model) for (id,model) in models], cam=cam, canvas_height=100, car_colors=car_colors)
end
```

```julia
using Reel
Reel.set_output_type("gif")

frames = Frames(MIME("image/png"), fps=10)
for frame_index in 1 : nframes(list_record)
    AutoMobius.unpack!(list_record, frame_index, simscene, models)
    
    # adjust target car to be at center
    del_s = simscene[target_ind].state.s + 100.0
    for i in 1 : length(simscene)
        veh = simscene[i]
        s′ = mod_position_to_roadway(veh.state.s - del_s, roadway)
        simscene[i] = Entity(PosSpeed1D(s′, veh.state.v), veh.def, veh.id)
    end
    
    frame = render(simscene, roadway,[AttentionOverlay(id, model) for (id,model) in models], cam=cam, canvas_height=100, car_colors=car_colors)
    push!(frames, frame)
end
    
frames
```

Unpacks the joint save into a list record.

```julia
frame_index = 1
simscene = MobiusScene()
simrec = QueueRecord(MobiusEntity, 100, get_timestep(rec))

AutoMobius.unpack!(list_record, frame_index, simscene, models)
does_sim_produce_collision(simscene, roadway, models, simrec, 2)
```

Generate a risk plot.

```julia
arr_time = collect(1:nframes(list_record)-1).*get_timestep(list_record)

prior = 0.1
nsimulations = 100
nticks = 100
betas_arr = [get_collision_estimation(list_record, j, roadway, prior, nsimulations=nsimulations, nticks=nticks) for j in 1  : nframes(list_record)-1]

arr_col_prob_means = [mean(β) for β in betas_arr]
arr_col_prob_q95 = [quantile(β,0.95) for β in betas_arr]
arr_col_prob_q05 = [quantile(β,0.05) for β in betas_arr]

Axis([Plots.Linear(arr_time, arr_col_prob_means, style="solid, thick, black, mark=none"),
      Plots.Linear(arr_time, arr_col_prob_q95, style="solid, black, mark=none"),
      Plots.Linear(arr_time, arr_col_prob_q05, style="solid, black, mark=none"),
      Plots.Linear([arr_time[end],arr_time[end]],[0,1.15], style="solid, black, mark=none"),
      Plots.Node(L"t_\text{collision}",arr_time[end],1.2,)        
     ],
    xlabel="time [s]", ylabel="probability of collision", ymin=0, ymax=1.3, width="15cm", height="10cm")
```
