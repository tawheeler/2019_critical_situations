```julia
using PGFPlots
include("system_optimization.jl")
```

```julia
f(x; a=1, b=5) = ((a-x[1])^2 + b*(x[2] - x[1]^2)^2 ) / 100

xdomain = (-2, 2)
ydomain = (-2, 2)

X = Vector{Float64}[]
for x₁ in linspace(xdomain...,21)
    for x₂ in linspace(ydomain...,21)
        push!(X, [x₁, x₂])
    end
end

plots = Plots.Plot[]
push!(plots, Plots.Contour(f, xdomain, ydomain, levels=[1,2,3,5,10,20,50,100], style="width=\\columnwidth", xbins=100, ybins=100))
push!(plots, Plots.Scatter([x[1] for x in X], [x[2] for x in X], style="mark=*, mark size=1, mark options={draw=black, opacity=0.4, fill=black}"))
Axis(plots, width="8cm", height="8cm", xlabel=L"x_1", ylabel=L"x_2", style="contour/labels=false, axis equal, view={0}{90}")
```

```julia
GP = GaussianProcess(m = x -> 200.0, k = (x,x′)->exp(-0.5*(norm(x-x′))^2), ν=0.0)
n = length(X)
μₚ = Array{Float64}(n)
νₚ = Array{Float64}(n)
sampled = falses(n)

function Base.push!(GP::GaussianProcess, i::Int)
    x = X[i]
    y = f(x)
    sampled[i] = true
    push!(GP, x, y)
end

srand(2)
push!(GP, rand(1:n))
push!(GP, rand(1:n))
predict!(μₚ, νₚ, GP, X)

best_index = select_sample_index_via_expected_improvement(GP, μₚ, νₚ, sampled)
x = X[best_index]
y_min = minimum(GP.y)
println(best_index, "  ", x)

g = GroupPlot(2,1, groupStyle="horizontal sep=4cm")
plots = Plots.Plot[]
push!(plots, Plots.Image((x,y)->predict(GP, [x,y])[1], xdomain, ydomain, style="width=\\columnwidth", xbins=100, ybins=100))
push!(plots, Plots.Contour(f, xdomain, ydomain, style="draw=white, width=\\columnwidth", xbins=100, ybins=100))
push!(plots, Plots.Scatter([p[1] for p in GP.X], [p[2] for p in GP.X], style="mark=o, mark size=1.5, mark options={draw=black}"))
push!(plots, Plots.Scatter([x[1]], [x[2]], style="mark=*, mark size=1, mark options={draw=red, fill=red}"))
push!(g, Axis(plots, width="8cm", height="8cm", xlabel=L"x_1", ylabel=L"x_2", title="GP mean", style="contour/labels=false, axis equal, view={0}{90}",
        xmin=xdomain[1], xmax=xdomain[2], ymin=ydomain[1], ymax=ydomain[2]))
plots = Plots.Plot[]
push!(plots, Plots.Image((x,y)->expected_improvement(y_min, predict(GP, [x,y])...), xdomain, ydomain, style="width=\\columnwidth", xbins=100, ybins=100))
push!(plots, Plots.Scatter([x[1]], [x[2]], style="mark=*, mark size=1, mark options={draw=red, fill=red}"))
push!(g, Axis(plots, width="8cm", height="8cm", xlabel=L"x_1", ylabel=L"x_2", title="expected improvement", style="contour/labels=false, axis equal, view={0}{90}", 
              xmin=xdomain[1], xmax=xdomain[2], ymin=ydomain[1], ymax=ydomain[2]))
g
```

```julia
for i in 1 : 50
    push!(GP, best_index)
    predict!(μₚ, νₚ, GP, X)

    best_index = select_sample_index_via_expected_improvement(GP, μₚ, νₚ, sampled)
    x = X[best_index]
    y_min = minimum(GP.y)
    println(best_index, "  ", x, "  ", f(x), "   ", y_min)
end

g = GroupPlot(2,1, groupStyle="horizontal sep=2cm")
plots = Plots.Plot[]
push!(plots, Plots.Image((x,y)->predict(GP, [x,y])[1], xdomain, ydomain, style="width=\\columnwidth", xbins=100, ybins=100))
push!(plots, Plots.Contour(f, xdomain, ydomain, style="draw=white, width=\\columnwidth", xbins=100, ybins=100))
push!(plots, Plots.Scatter([p[1] for p in GP.X], [p[2] for p in GP.X], style="mark=o, mark size=1.5, mark options={draw=magenta}"))
push!(plots, Plots.Scatter([x[1]], [x[2]], style="mark=*, mark size=1, mark options={draw=red, fill=red}"))
push!(g, Axis(plots, width="8cm", height="8cm", xlabel=L"x_1", ylabel=L"x_2", title="GP mean", style="contour/labels=false, axis equal, view={0}{90}",
        xmin=xdomain[1], xmax=xdomain[2], ymin=ydomain[1], ymax=ydomain[2]))
plots = Plots.Plot[]
push!(plots, Plots.Image((x,y)->expected_improvement(y_min, predict(GP, [x,y])...), xdomain, ydomain, style="width=\\columnwidth", xbins=100, ybins=100))
push!(plots, Plots.Scatter([x[1]], [x[2]], style="mark=*, mark size=1, mark options={draw=red, fill=red}"))
push!(g, Axis(plots, width="8cm", height="8cm", xlabel=L"x_1", ylabel=L"x_2", title="expected improvement", style="contour/labels=false, axis equal, view={0}{90}", 
              xmin=xdomain[1], xmax=xdomain[2], ymin=ydomain[1], ymax=ydomain[2]))
g
```
